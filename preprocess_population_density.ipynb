```python
import pandas as pd

density_df = pd.read_csv("drive/MyDrive/DnA/인구밀도_전처리전.csv", encoding="utf-8-sig", header=1)

density_df = density_df[['동별(2)', '인구밀도 (명/㎢)']]

density_df.columns = ['구', '인구밀도']

density_df = density_df.dropna(subset=['구', '인구밀도'])
density_df = density_df[density_df['구'] != '소계']
density_df = density_df[density_df['인구밀도'].astype(str).str.isnumeric()]
density_df['인구밀도'] = density_df['인구밀도'].astype(int)

print("==== 전처리 후 데이터 ====")
print(density_df)

print("\n==== 데이터 기본 정보 ====")
density_df.info()

print("\n==== 컬럼명 확인 ====")
print(density_df.columns)

density_df.to_csv("drive/MyDrive/DnA/인구밀도_전처리후.csv", index=False, encoding="utf-8-sig")
print("\n'인구밀도_전처리후.csv' 파일이 생성되었습니다.")
```

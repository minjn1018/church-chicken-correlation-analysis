```python
import pandas as pd
from io import TextIOWrapper
from google.colab import drive

drive.mount('/content/drive')

with open("/content/drive/MyDrive/DnA/소득_전처리전.csv", "rb") as f:
    wrapper = TextIOWrapper(f, encoding="cp949", errors="ignore")
    salary_df = pd.read_csv(wrapper)

print("==== 전처리 전 데이터 미리보기 ====")
print(salary_df.head())

print("\n==== 데이터 기본 정보 ====")
salary_df.info()

print("\n==== 컬럼명 ====")
print(salary_df.columns)
```

```python
salary_df = salary_df[['행정동_코드_명', '월_평균_소득_금액']]

salary_df['구'] = salary_df['행정동_코드_명'].str.extract(r'(\S+구)')

salary_df = salary_df.dropna(subset=['구', '월_평균_소득_금액'])

salary_grouped = salary_df.groupby('구', as_index=False)['월_평균_소득_금액'].mean()

salary_grouped['월_평균_소득_금액'] = salary_grouped['월_평균_소득_금액'].astype(int)

print("\n==== 최종 전처리된 데이터 미리보기 ====")
print(salary_grouped)

print("\n==== 컬럼명 ====")
print(salary_grouped.columns.tolist())

salary_grouped.to_csv("/content/drive/MyDrive/DnA/소득_전처리후.csv", index=False, encoding="utf-8-sig")
print("\n'인구밀도_전처리후.csv' 파일이 생성되었습니다.")
```

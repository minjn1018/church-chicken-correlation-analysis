```python
import pandas as pd
from google.colab import drive
from io import TextIOWrapper

drive.mount('/content/drive')

with open("drive/MyDrive/DnA/서울시 일반음식점 인허가 정보_전처리전.csv", "rb") as f:
    wrapper = TextIOWrapper(f, encoding="cp949", errors="ignore")
    chicken_df = pd.read_csv(wrapper, low_memory=False)

print("==== 전처리 전 데이터 ====")
print(chicken_df.head())

print("\n==== 데이터 기본 정보 ====")
chicken_df.info()

print("\n==== 컬럼명 확인 ====")
print(chicken_df.columns)
```

```python
chicken_active_df = chicken_df[
    chicken_df['위생업태명'].str.contains('치킨|통닭', na=False)
    & (chicken_df['영업상태명'] == '영업/정상')
]

chicken_active_df['구'] = (
    chicken_active_df['도로명주소'].str.extract(r'(\S+구)')[0]
        .fillna(chicken_active_df['지번주소'].str.extract(r'(\S+구)')[0])
)

chicken_count = chicken_active_df['구'].value_counts().reset_index()
chicken_count.columns = ['구', '치킨집수']

print("==== 구별 치킨집 수 ====")
print(chicken_count.head(10))

chicken_active_df.to_csv(
    "drive/MyDrive/DnA/서울시 일반음식점 인허가 정보_전처리후.csv",
    index=False,
    encoding="cp949"
)
print("\n'서울시 일반음식점 인허가 정보_전처리후.csv' 파일이 생성되었습니다.")
```

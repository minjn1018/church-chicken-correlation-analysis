```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

chicken_df = pd.read_csv("drive/MyDrive/DnA/서울시 일반음식점 인허가 정보_전처리후.csv", encoding="cp949")
density_df = pd.read_csv("drive/MyDrive/DnA/인구밀도_전처리후.csv", encoding="utf-8-sig")
salary_df = pd.read_csv("drive/MyDrive/DnA/소득_전처리후.csv", encoding="utf-8-sig")

chicken_df = chicken_df[
    chicken_df['위생업태명'].str.contains('통닭|치킨', na=False)
    & (chicken_df['영업상태명'] == '영업/정상')
]

chicken_df['구'] = (
    chicken_df['도로명주소'].str.extract(r'(\S+구)')[0]
        .fillna(chicken_df['지번주소'].str.extract(r'(\S+구)')[0])
)

chicken_count = chicken_df['구'].value_counts().reset_index()
chicken_count.columns = ['구', '치킨집수']

merged_df = pd.merge(chicken_count, density_df, on='구', how='inner')

final_df = pd.merge(merged_df, salary_df, on='구', how='inner')

final_df.to_csv("drive/MyDrive/DnA/서울시_치킨_인구밀도_소득_병합.csv", index=False, encoding="utf-8-sig")

print(final_df.head(10))
print("\n병합 데이터가 저장되었습니다!")
```
